```python
import pandas as pd
from lightgbm import LGBMRegressor
from sklearn.datasets import fetch_openml
```

```python
pd.set_option("display.max_rows", 500)
```

```python
df = fetch_openml(data_id=546)
X = df.data
y = df.target
```

```python
model = LGBMRegressor(
    n_estimators=10,
    max_depth=1,
    learning_rate=0.1,
    max_cat_to_onehot=1,
    cat_l2=0.0,
    cat_smooth=0.0,
    min_data_per_group=1,
    max_cat_threshold=1000,
)
model.fit(X, y)
```

```python
list(X.columns)
```

```python
df = model.booster_.trees_to_dataframe()
df_mod = df.loc[df["weight"] == 0]
df_mod.head(10)
```
